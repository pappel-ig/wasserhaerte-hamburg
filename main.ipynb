```python
import requests
import json
```

```python
plz_wasser_info = {}
for plz in open("plz.txt"):
    plz = plz.strip()
    plz = plz.replace("\n", "")
    response = requests.get('https://www.hamburgwasser.de/startseite/waterworkApi', {
        'tx_project_waterwork[zip]': plz
    })
    body = json.loads(response.content)
    if response.status_code == 200:
        print(plz)
        if not body['streets']:
            if len(body['waterworks']) > 0:
                if len(body['waterworks']) > 1:
                    print("bad")
                plz_wasser_info[plz] = body['waterworks'][0]
        else:
            plz_wasser_info[plz] = []
    else:
        print(response.status_code)
        break
plz_wasser_info
```

```python
with open('pre_processed_0.json', 'w') as outfile:
    outfile.write(json.dumps(plz_wasser_info))
```

```python
with open('pre_processed_0.json', 'r') as infile:
    plz_wasser_info = json.loads(infile.read())

for plz in plz_wasser_info:
    if plz_wasser_info[plz]:
        plz_wasser_info[plz] = plz_wasser_info[plz]['watervalues'][4]['value']
        
plz_wasser_info
```

```python
with open('pre_processed_1.json', 'w') as outfile:
    outfile.write(json.dumps(plz_wasser_info))
```

```python
with open('pre_processed_1.json', 'r') as infile:
    plz_wasser_info = json.loads(infile.read())
    
for plz in plz_wasser_info:
    if not plz_wasser_info[plz]:
        print(plz)
        response = requests.get('https://www.hamburgwasser.de/startseite/waterworkApi', {
            'tx_project_waterwork[zip]': plz
        })
        body = json.loads(response.content)
        elem = {}
        for street in body['streets']:
            elem[street] = {}
        plz_wasser_info[plz] = elem
plz_wasser_info
```

```python
with open('pre_processed_2.json', 'w') as outfile:
    outfile.write(json.dumps(plz_wasser_info))
```

```python
with open('pre_processed_2.json', 'r') as infile:
    plz_wasser_info = json.loads(infile.read())
    
for plz in plz_wasser_info:
    if type(plz_wasser_info[plz]) == dict:
        print("### " + plz + " ###")
        for street in plz_wasser_info[plz]:
            print(street)
            response = requests.get('https://www.hamburgwasser.de/startseite/waterworkApi', {
                'tx_project_waterwork[zip]': plz,
                'tx_project_waterwork[street]': street
            })
            body = json.loads(response.content)
            plz_wasser_info[plz][street] = body['waterworks'][0]['watervalues'][4]['value']
plz_wasser_info
        
```

```python
with open('pre_processed_3.json', 'w') as outfile:
    outfile.write(json.dumps(plz_wasser_info))
```

```python
with open('pre_processed_3.json', 'r') as infile:
    plz_wasser_info = json.loads(infile.read())

for plz in plz_wasser_info:
    if type(plz_wasser_info[plz]) == dict:
        sum = 0
        for street in plz_wasser_info[plz]:
            sum = sum + float(str(plz_wasser_info[plz][street]).replace(',', '.'))
        plz_wasser_info[plz] = round(sum/len(plz_wasser_info[plz]), 1)
    else:
        plz_wasser_info[plz] = float(str(plz_wasser_info[plz]).replace(',', '.'))
plz_wasser_info
```

```python
import geopandas as gpd

plz_geo = gpd.read_file('https://opendata.rhein-kreis-neuss.de/api/v2/catalog/datasets/nrw-postleitzahlen/exports/geojson')
plz_geo
```

```python
import pandas as pd

with open('pre_processed_4.json', 'r') as infile:
    plz_wasser_info = json.loads(infile.read())
    
counter = 0
data = {}
for plz in plz_wasser_info:
    data[counter] = [plz, plz_wasser_info[plz]]
    counter += 1
    
df = pd.DataFrame.from_dict(data).transpose()
df.rename(columns={ df.columns[0]: 'plz', df.columns[1]: 'haertegrade' }, inplace = True)
df
```

```python
plz_geo['plz'] = plz_geo['name'].astype(str)
df['PLZ'] = df['plz'].astype(str)

merged = plz_geo.merge(df, left_on='name', right_on='PLZ')
merged
```

```python
import folium

m = folium.Map(location=[53.550556, 9.993333], zoom_start=10)

# Choropleth-Layer hinzufügen
folium.Choropleth(
    geo_data=merged,  # Verwende die zusammengeführten GeoJSON-Daten
    data=merged,  # Daten, die visualisiert werden sollen
    fill_color='Purples',
    key_on='feature.properties.PLZ',
    columns=['PLZ', 'haertegrade'],
).add_to(m)
m
```
